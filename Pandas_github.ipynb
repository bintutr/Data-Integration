# **ACTIVIDAD 2**

*Creamos un DataFrame utilizando el fichero de la Url.*

```python
import pandas as pd
df_url = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/Cars93_miss.csv')                     
```

* Hacemos una pequeña revisión de la estructura del Dataframe

```python
print(df_url)
```

## 1. Revisión valores nulos en el Dataframe. El comando que se utilice para este propósito deberá devolver únicamente un resultado binario, es decir, si hay o no presente nulos en el Dataframe

```python
Valores_nulos = df_url.isnull().values.any()
print(Valores_nulos)
```

Revisión de valores nulos. Si es TRUE si que hay valores nulos, si es FALSE, no hay valores nulos.

## 2. Contar el número de nulos en cada columna

```python
df_url.isnull().sum()
```

Número de valores nulos en cada columna

## 3. ¿Cuál es el valor máximo de los precios (Columna ‘Price’) de un carro?

```python
max_Price = df_url['Price'].max()
print(max_Price)
```

Valor máximo de la columna 'Price' es 61.9

## 4. Obtener, en un solo paso, los valores del fabricante, modelo y tipo ('Manufacturer','Model' y 'Type') del carro de mayor precio (obtenido en el apartado 3)

```python
carro_mayor_precio = df_url.loc[df_url['Price'].idxmax(), ['Manufacturer', 'Model', 'Type']]

print(carro_mayor_precio)
```

Se muestran los valores de las columnas `'Manufacturer'`, `'Model'`,`'Type'` en función del valor máximo de la columna `'Price'`.
